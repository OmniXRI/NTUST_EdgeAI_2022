# 安裝OpenVINO 2021版在Colab上

OpenVINO可支援APT方式安裝（如下參考資料來源），但在Colab上須略微調整，包括：   
1.   去掉sudo
2.   指定對應Ubuntu版本將下載指令加入安裝清單中
3.   加入清單方式從tee改成 >>

**本範例僅限OpenVINO 2021.x.x版**  

參考資料來源：https://docs.openvino.ai/2021.4/openvino_docs_install_guides_installing_openvino_apt.html  
歐尼克斯實境互動工作室 OmniXRI Jack, 2022.04.29 整理製作

```python
# 查詢系統Ubuntu版本(Colab預設為 Ubuntu 18.04.5 LTS，本步驟可略過)
!lsb_release -a 
```

```python
# 取得OpenVINO 2021公開金錀
!wget https://apt.repos.intel.com/openvino/2021/GPG-PUB-KEY-INTEL-OPENVINO-2021 
# 加入OpenVINO公開金錀到系統金錀群中
!apt-key add GPG-PUB-KEY-INTEL-OPENVINO-2021 
# 建立更新安裝清單檔案
!touch /etc/apt/sources.list.d/intel-openvino-2021.list
# 將下載指令加入安裝清單中
!echo "deb https://apt.repos.intel.com/openvino/2021 all main" >> /etc/apt/sources.list.d/intel-openvino-2021.list
# 更新系統
!apt update
```

```python
# 查詢可使用OpenVINO版本（本動作為檢查可用版本，本步驟可略過）
# 查詢動作必須先完成上述指令後才能查詢
!apt-cache search intel-openvino 
```

```python
# 安裝OpenVINO開發者指定版本到虛擬機系統中（例2021.4.752版）
!apt install intel-openvino-dev-ubuntu18-2021.4.752 
```

```python
# 檢查OpenVINO是否已安裝到指定路徑，本步驟可略過
!ls /opt/intel
```
